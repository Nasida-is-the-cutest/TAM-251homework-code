```python
#HW4.7. Maximum Shear Stress in Composite Shaft
l=44
ga=4000
gs=10000
da=5
ds=3
tb=1

js=np.pi/2*(ds/2)**4
ja=np.pi/2*((da/2)**4-(ds/2)**4)

phi=tb/(ja*ga/l+js*gs/l)

ts=phi*js*gs/l
ta=phi*ja*ga/l

tao_s=ts*(ds/2)/js
tao_a=ta*(da/2)/ja

print(tao_a)
print(tao_s)
```
