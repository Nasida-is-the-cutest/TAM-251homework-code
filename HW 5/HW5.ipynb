```python
#HW5.1. Torsion - pipe around solid rod - angle of twist
import numpy as np 
l=46
ga=3000
gs=11000
da=5
ds=4
tb=4

js=np.pi/2*(ds/2)**4
ja=np.pi/2*((da/2)**4-(ds/2)**4)

phi=tb/(ja*ga/l+js*gs/l)
print(phi)
```

```python
#HW5.2. Torsion - solid+hollow shaft - diameter
di=37
tb=1535
tc=449
g=118
l=1051
tao_max=11

di=di*1e-3
l=l*1e-3
g=g*1e6
tao_max=tao_max*1e6

t1=tb-tc
t2=tc
print(t1)
print(-t2)

from scipy.optimize import fsolve
def f(d):
    return t1*d/2-tao_max*np.pi/2*(d/2)**4
d_tao=fsolve(f,0.1)[0]
print(d_tao*1000)


j1=np.pi/2*(d_tao/2)**4
j2=np.pi/2*((d_tao/2)**4-(di/2)**4)
phi1=t1*l/j1/g
phi2=t2*l/j2/g
phi=abs(abs(phi1)-abs(phi2))
print(phi)
```

```python
#HW5.3. Fixed End Steel Shaft
d=40
a=53
g=75
l1=0.4
l2=0.6

d=d*1e-3
a=a*1e-3
g=g*1e9
f=3000
t=2*f*a
j=np.pi/2*(d/2)**4
phi=t/j/g/(1/l1+1/l2)
t1=phi*j*g/l1
t2=phi*j*g/l2
tao1=t1*d/2/j
tao2=t2*d/2/j
print(tao1/1000000)
print(tao2/1000000)
```

```python
#HW5.4. Torque on Two Shafts System

g=11e3*1e3
d=1
t=79*12 #英制单位*****
r1=6
r2=4
t2=40*12 #AB上的力矩

c=d/2
j=np.pi/2*c**4
t_gear=t2/r2*r1
td=t_gear-t
phi_1=td*10/j/g+(t+td)*30/j/g #DC最右侧的点的Φ

phi_2=phi_1*r1/r2
phi_3=t2*10/j/g

print(np.degrees(phi_2+phi_3))
```

```python
#HW5.6. Gears – Three Rods – Twist Angle
r1=156
r2=45
d=32
l=130
g=73
te=423

r1=r1*1e-3
r2=r2*1e-3
d=d*1e-3
l=l*1e-3
g=g*1e9

j=np.pi/2*(d/2)**4
t1=te
t2=r2/r1*te
t3=(r2/r1)**2*te #最上面

phi3=t3*l/j/g
phi3_2=phi3*r2/r1
phi2=t2*l/j/g
phi2_1=(phi3_2+phi2)*r2/r1
phi1=t1*l/j/g

phi=phi1+phi2_1
print(phi)
```

```python
#HW5.7. Maximum Shear Stress in Hollow Shaft System
g=67
d1=34
d2=58
tb=1278
tc=1559
l1=259
l2=573
l3=272

g=g*1e9
d1=d1*1e-3
d2=d2*1e-3
l1=l1*1e-3
l2=l2*1e-3
l3=l3*1e-3

j1=np.pi/2*(d1/2)**4
j2=np.pi/2*(d2/2)**4
j3=np.pi/2*((d2/2)**4-(d1/2)**4)
from scipy.optimize import fsolve
def f(ta):
    tab=ta
    tbc=ta-tb
    tcd=ta-tb-tc
    return tab*l1/j1+tbc*l2/j2+tcd*l3/j3
ta=fsolve(f,100)[0]
tao1=ta*d1/2/j1
tao2=abs(ta-tb)*d2/2/j2
tao3=abs(ta-tb-tc)*d2/2/j3
print(tao1/1000000)
print(tao2/1000000)
print(tao3/1000000)
```

```python
#HW5.8. Maximum Shear Stress in Two Segments Shaft
t=690
d=30e-3
gst=75e9
gbr=39e9

jst=np.pi/2*(d/2)**4
phi=t/2*0.75/jst/gst
dbr=(t/2*1.6/gbr/phi*2/np.pi)**0.25*2
print(dbr*1000)
jbr=np.pi/2*(dbr/2)**4
tao_br=t/2*dbr/2/jbr/1000000
tao_st=t/2*d/2/jst/1000000
print(max(tao_br,tao_st))
```
