```python
#HW3.3. Yield Strength of Bone
#0.3% offset 的定义是：画一条与弹性初始段斜率相同、应变轴截距为 0.003的直线，该直线与应力 - 应变曲线的交点即为屈服强度。
#弹性段的斜率位x=0处的导数
from scipy.optimize import fsolve
epsilon = 0.004

k=0.45*10**(-6)

def f(x):
    return k*x+epsilon-(0.45*x*10**(-6)+0.36*x**3*10**(-12))
sigma_ys=fsolve(f, 2)[0]
print(sigma_ys/1000)

```

```python
#HW3.7. Elongation - two rods attached to rigid bar
k1 = 235 #N/mm
L=219 #mm
theta = np.radians(39)
k2= 152 #N/mm
P=24 #N

import numpy as np
from sympy import *
t = symbols('t') #设t表示刚性梁绕 B 点的小转动角度
d1 = ((L*sin(theta))**2 + (L*cos(theta) + t*L)**2)**0.5 - L
d2 = t*L
F1 = k1*d1
F2 = k2*d2
eq1 = P*L/2 - F2*L - F1*cos(theta)*L 
print(solve(eq1, t))
solve(eq1, t)[1]*L
```

```python
#HW3.10. Vertical Displacement of A Point
l=1010*10**(-3) #mm
b=531*10**(-3) #mm
e = 192*10**9 #Pa
ac = 865*10**(-6) #mm^2
p = 1000 #N
a = 534*10**(-3) #mm
theta = np.radians(59)

x=symbols('x')
delta_b = a/l*x
delta_bd=((b*np.cos(theta))**2+(delta_b+np.sin(theta))**2)**0.5-b
p_bd=ac*e*delta_bd/l
eq=p*l-p_bd*np.sin(theta)*a
print(solve(eq, x))
```

```python
L = 1039 # mm
b = 518 # mm
E = 190 # GPa
P = 1 # kN
a = 517 # mm
Ac = 865 # mm^2
theta = 60 # degrees

#单位转化
P=P*1e3
a=a*1e-3
b=b*1e-3
L=L*1e-3
E=E*1e9
Ac=Ac*1e-3*1e-3
theta=theta*pi/180

alpha=Symbol('alpha')

#拉伸模量 E= 正应力/正应变
#正应力 σ=力/面积

#计算杆的力（力矩平衡）
Fbd=P*L/a/sin(theta)

#弹性模数→弹性系数
k=E*Ac/b

#计算形变量←正应变
dL=Fbd/k #实际形变

#近似形变
dx=alpha*a*sin(theta)

alpha=solve([dx-dL],[alpha])[alpha]

float(L*alpha*1e3)
```
