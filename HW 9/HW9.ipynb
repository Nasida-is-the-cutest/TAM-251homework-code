```python
#HW9.1. Determine stresses in rotated system of coordinates
import numpy as np

sigma_x = -17.0 # MPa
sigma_y = 27.0 # MPa
tau_xy = 15.0 # MPa
theta=np.radians(35)

sigma_xp=(sigma_x+sigma_y)/2+(sigma_x-sigma_y)/2*np.cos(2*theta)+tau_xy*np.sin(2*theta)
sigma_yp=(sigma_x+sigma_y)/2-(sigma_x-sigma_y)/2*np.cos(2*theta)-tau_xy*np.sin(2*theta)
tau_xyp=-(sigma_x-sigma_y)/2*np.sin(2*theta)+tau_xy*np.cos(2*theta)

print(sigma_xp)
print(sigma_yp)
print(tau_xyp)
```

```python
#HW9.2. Obtain principal stresses
sigma_x = 19.0 # MPa
sigma_y = 0 # MPa
tau_xy = 20.0 # MPa

sigma1=(sigma_x+sigma_y)/2+(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5
sigma2=(sigma_x+sigma_y)/2-(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5

theta_p=np.degrees(np.arctan(tau_xy*2/(sigma_x-sigma_y))/2)

tau_max=(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5
print(sigma1)
print(sigma2)
print(theta_p)
print(tau_max)
```

```python
#HW9.3. Combined Loading: stresses and failure
import numpy as np

p = 1432 # kPa
F = 12 # kN
t = 8 # mm
r = 136 # mm
sigmaY = 218 # MPa

p=p*1e3
f=F*1e3
t=t*1e-3
r=r*1e-3
sigmaY=sigmaY*1e6

sigma_y=p*r/t
sigma_x=f/(2*np.pi*r*t)+r*p/2/t
print("sigma_x = ",sigma_x/1e6)
print("sigma_y = ",sigma_y/1e6)
tau_xy=0
print("tau_xy  = ",tau_xy)

sigma1=(sigma_x+sigma_y)/2+(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5
sigma2=(sigma_x+sigma_y)/2-(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5
theta_p=np.degrees(np.arctan(tau_xy*2/(sigma_x-sigma_y))/2)+90

print("sigma1 = ",sigma1/1e6)
print("sigma2 = ",sigma2/1e6)
print("theta_p = ",theta_p)

sigma_von=(sigma1**2+sigma2**2-sigma1*sigma2)**0.5
FS=sigmaY/sigma_von
print("FS = ",FS)
```

```python
#HW9.4. Combined Loading: stresses and failure
import numpy as np

p = 1432 # kPa
T = 1 # kN.m
t = 18 # mm
r = 324 # mm
sigmaY = 250 # MPa

p=p*1e3
t=t*1e-3
r=r*1e-3
T=T*1e3
sigmaY=sigmaY*1e6

sigma_y=p*r/t
sigma_x=r*p/2/t
print("sigma_x = ",sigma_x/1e6)
print("sigma_y = ",sigma_y/1e6)
R=r+t
j=np.pi/2*(R**4-r**4)
tau_xy=-T*R/j  
print("tau_xy  = ",tau_xy/1e6)

sigma1=(sigma_x+sigma_y)/2+(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5
sigma2=(sigma_x+sigma_y)/2-(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5
theta_p=np.degrees(np.arctan(tau_xy*2/(sigma_x-sigma_y))/2)+90

print("sigma1 = ",sigma1/1e6)
print("sigma2 = ",sigma2/1e6)
print("theta_p = ",theta_p-180)

tau_max=(((sigma_x-sigma_y)/2)**2+tau_xy**2)**0.5
print("tau_max = ",tau_max/1e6)
sigma_max=sigma1
FS=sigmaY/sigma_max
print("FS = ",FS)


#-90<=theta_p<=90
```

```python
#HW9.6. Find normal and shear stresses in pressure vessel weld plane
import numpy as np

alpha = 59 # degrees
r = 700 # mm
t = 15 # mm
p = 2.6 # MPa

alpha=np.radians(alpha)
r=r*1e-3
t=t*1e-3
p=p*1e6
theta=np.pi/2-alpha

sigma_y=p*r/t
sigma_x=r*p/2/t
print("sigma_x = ",sigma_x/1e6)
print("sigma_y = ",sigma_y/1e6)

tau_xy = 0 # MPa
print("tau_xy = ", tau_xy)
sigma_xp=(sigma_x+sigma_y)/2+(sigma_x-sigma_y)/2*np.cos(2*theta)+tau_xy*np.sin(2*theta)
sigma_yp=(sigma_x+sigma_y)/2-(sigma_x-sigma_y)/2*np.cos(2*theta)-tau_xy*np.sin(2*theta)
tau_xyp=-(sigma_x-sigma_y)/2*np.sin(2*theta)+tau_xy*np.cos(2*theta)

print("sigma_xp = ",sigma_xp/1e6)
print("sigma_yp = ",sigma_yp/1e6)
print("tau_xyp = ",tau_xyp/1e6)
```
