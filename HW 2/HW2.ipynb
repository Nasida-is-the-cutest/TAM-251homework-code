```python
#HW2.1. Normal Strain in a rectangle
a=536
b=400
dd=3
db=4
l_initial=(a**2+b**2)**0.5
l_final=((a-db)**2+(b-dd)**2)**0.5
epsilon=(l_final-l_initial)/l_initial
print("The normal strain is: ", epsilon*1000)
```

```python
#HW2.3. Average Shear Strain at Corner
a=4.2
ab=400
ad=300
ad2=(ad**2+a**2)**0.5
ab2=(ab**2+a**2)**0.5
l=((ab-a)**2+(ad-a)**2)**0.5
angle=np.arccos((ad2**2+ab2**2-l**2)/(2*ad2*ab2))
gamma=np.pi/2-angle
print("The shear strain at the corner is: ", gamma)
```

```python
#HW2.4. Shear Strain of Rectangular Piece
x=386
y=307

c2d=np.array([-2,-2-y]) #C2到D的向量
c2b2=np.array([x+3,2]) #C2到B2的向量
norm_c2d=np.linalg.norm(c2d)
norm_c2b2=np.linalg.norm(c2b2)
cos_angle_c=np.dot(c2d,c2b2)/(norm_c2d*norm_c2b2)
angle_c=np.arccos(cos_angle_c)
gamma_c=np.pi/2-angle_c
print("The shear strain of the rectangular piece gamma_c is: ", gamma_c)

dc2=np.array([2,y+2]) #C新坐标位C2
da2=np.array([x+3,2]) #A新坐标位A2

norm_da2=np.linalg.norm(da2)
norm_dc2=np.linalg.norm(dc2)
cos_angle_d=np.dot(da2,dc2)/(norm_da2*norm_dc2)
angle_d=np.arccos(cos_angle_d)
gamma_d=np.pi/2-angle_d
print(angle_d)
print("The shear strain of the rectangular piece gamma_d is: ", gamma_d)
```

```python
#HW2.5. Average Shear Strain at Point
epsilon_xy=0.007
theta=0.007

f2=np.array([600*np.sin(theta),600*np.cos(theta)]) #点f的变形后坐标
g=np.array([300,0]) #点g的坐标
c2=np.array([900+300*np.sin(theta),300*np.cos(theta)]) #点c的变形后坐标

gf2=g-f2
gc2=g-c2
norm_gf2=np.linalg.norm(gf2)
norm_gc2=np.linalg.norm(gc2)
cos_angle_g=np.dot(gf2,gc2)/(norm_gf2*norm_gc2)
angle_g=np.arccos(cos_angle_g)
gamma_g=np.pi/2-angle_g
print("The average shear strain at point g is: ", gamma_g)
```

```python
#HW2.6. Deformation of A Rod
l=251 #mm
d1=18  #mm
g=117 #GPa
p=115000 #N
delta=0.3 #mm
sigma_y=301 #MPa
d=d1/1000 #mm转m

sigma=p/(np.pi*(d/2)**2) #计算应力
E=sigma*l/delta #计算杨氏模量
E=E/1e9 #Pa转GPa
print("The Young's modulus of the rod is: ", E, "GPa")

v=E/(2*g)-1 #计算泊松比
print("The Poisson's ratio of the rod is: ", v)


delta_d=-v*delta/l*d1
print("The change in diameter of the rod is: ", delta_d, "mm")
```
