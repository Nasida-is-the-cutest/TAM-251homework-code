```python
#HW8.4. Eccentric loading of a beam
from sympy import *
import numpy as np

b = 217.00 # mm
t = 69.00 # mm
P = 77.00 # kN

b=b/1e3
t=t/1e3
p=P*1000

sigma1=p/(t*b)
d=S('d')
m=(b/2-d)*p
I=t*b**3/12
sigma2=m*b/2/I            #c=b/2
solution=solve(sigma2-sigma1)[0]
print(solution*1e3)
```

```python
#HW8.6. Combined Loading: determine state of plane stress
from sympy import *

# ===================== 1. 定义已知参数 + 单位统一（核心！）=====================
# 原始参数（你提供的）
p_kPa = 1302    # 内压，单位 kPa
T_kNm = 6       # 扭矩，单位 kN·m
V_kN = 9        # 剪力，单位 kN
L = 1020         # 长度，单位 mm
t = 17          # 壁厚，单位 mm
r = 170         # 内半径，单位 mm

# 单位转换（统一为：N, mm, MPa；1 MPa = 1 N/mm²）
p = p_kPa / 1000  # kPa → MPa（1 kPa = 0.001 MPa）
T = T_kNm * 10**6 # kN·m → N·mm（1 kN·m = 1e6 N·mm）
V = V_kN * 1000   # kN → N
R = r + t         # 外半径，mm

# ===================== 2. 分步计算应力 =====================
# ① 内压产生的应力
sigma_x1 = p * r / (2 * t)      # 轴向应力（内压）
sigma_y = p * r / t             # 环向应力（仅内压贡献）

# ② 弯矩产生的轴向弯曲应力
M = V * L                       # 弯矩 M = V·L
I = pi/4 * (R**4 - r**4)        # 环形截面惯性矩
sigma_x2 = (M * R) / I          # 弯曲应力

# ③ 总轴向应力
sigma_x = sigma_x1 - sigma_x2

# ④ 扭矩产生的剪应力
J = 2 * I                       # 极惯性矩（圆截面 J=2I）
tau_xy = -(T * R) / J           # 扭转剪应力（负号为符号规定）

# ===================== 3. 输出结果（保留4位小数） =====================
print("===== 点H的平面应力结果 =====")
print(f"轴向正应力 σx = {sigma_x.evalf():.4f} MPa")
print(f"环向正应力 σy = {sigma_y} MPa")
print(f"剪应力 τxy = {tau_xy.evalf():.4f} MPa")
```
